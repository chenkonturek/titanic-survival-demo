```python
import sys
sys.path.insert(0, '/Users/chen/PycharmProjects/claude-demo/titanic-survival-prediction/02_src')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from visualization import (
    plot_survival_by_gender,
    plot_survival_by_class,
    plot_age_distribution,
    plot_survival_by_family_size,
    plot_fare_distribution,
    plot_correlation_heatmap,
    plot_survival_by_title,
)

PROCESSED_PATH = '/Users/chen/PycharmProjects/claude-demo/titanic-survival-prediction/00_data/processed/titanic_cleaned.parquet'
FIGURES_DIR = '/Users/chen/PycharmProjects/claude-demo/titanic-survival-prediction/03_reports/figures'

df = pd.read_parquet(PROCESSED_PATH)
print(f"Dataset shape: {df.shape}")
df.head()
```

## 1. Dataset Overview

```python
print("Data types and missing values:")
print(df.info())
print("\nDescriptive statistics:")
df.describe()
```

```python
total = len(df)
survived = df['survived'].sum()
print(f"Total passengers: {total}")
print(f"Survived: {int(survived)} ({survived/total*100:.1f}%)")
print(f"Did not survive: {total - int(survived)} ({(total-survived)/total*100:.1f}%)")
```

## 2. Survival by Gender

Women had a dramatically higher survival rate (74.2%) compared to men (18.9%), reflecting the "women and children first" evacuation protocol.

```python
plot_survival_by_gender(df, f'{FIGURES_DIR}/survival_by_gender.png')
from IPython.display import Image
Image(f'{FIGURES_DIR}/survival_by_gender.png')
```

## 3. Survival by Passenger Class

1st class passengers had a 63% survival rate versus only 24% for 3rd class, suggesting strong socioeconomic privilege in lifeboat access.

```python
plot_survival_by_class(df, f'{FIGURES_DIR}/survival_by_class.png')
Image(f'{FIGURES_DIR}/survival_by_class.png')
```

## 4. Age Distribution by Survival

Young children show higher survival rates. The age distribution is broadly similar between survivors and non-survivors, though children under 10 have elevated survival odds.

```python
plot_age_distribution(df, f'{FIGURES_DIR}/age_distribution.png')
Image(f'{FIGURES_DIR}/age_distribution.png')
```

## 5. Survival by Family Size

Passengers traveling with small families (2–4 members) had higher survival rates than those traveling alone or in very large groups. Solo travelers had only a 30% survival rate versus 51% for those with family.

```python
plot_survival_by_family_size(df, f'{FIGURES_DIR}/survival_by_family_size.png')
Image(f'{FIGURES_DIR}/survival_by_family_size.png')
```

## 6. Fare Distribution by Survival

Survivors paid significantly higher fares (median £26) compared to non-survivors (median £10.50), reinforcing the socioeconomic dimension of survival.

```python
plot_fare_distribution(df, f'{FIGURES_DIR}/fare_distribution.png')
Image(f'{FIGURES_DIR}/fare_distribution.png')
```

## 7. Survival by Title

Title (a proxy for gender and social status) shows stark differences. Mrs and Miss categories have ~65–76% survival, while Mr has only 16.4%.

```python
plot_survival_by_title(df, f'{FIGURES_DIR}/survival_by_title.png')
Image(f'{FIGURES_DIR}/survival_by_title.png')
```

## 8. Feature Correlation Matrix

```python
plot_correlation_heatmap(df, f'{FIGURES_DIR}/correlation_heatmap.png')
Image(f'{FIGURES_DIR}/correlation_heatmap.png')
```

## 9. Key EDA Insights Summary

| Finding | Detail |
|---|---|
| Gender is the strongest predictor | Women survived at 74.2% vs. men at 18.9% |
| Passenger class strongly correlated with survival | 1st: 62.9%, 2nd: 47.3%, 3rd: 24.2% |
| Higher fare = higher survival | Survivors median fare £26 vs £10.50 for non-survivors |
| Traveling alone reduces survival odds | Solo: 30.4% vs with family: 50.6% |
| Title encodes gender + social status | Mrs (75.6%) and Miss (65.1%) greatly outperform Mr (16.4%) |
| Age has modest direct effect | Children under 10 benefit from "children first" protocol |

**Features most correlated with survival:** sex (0.54), pclass (-0.34), fare (0.26), Title (-0.54), IsAlone (-0.20)
